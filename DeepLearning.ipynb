```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# 1. Data Preparation
def load_and_preprocess_data(filepath):
    data = pd.read_csv(filepath, parse_dates=['dates'])
    
    # Create temporal features
    data['days_since_start'] = (data['dates'] - data['dates'].min()).dt.days
    data['month'] = data['dates'].dt.month
    data["Hs_ma_7"] = data['Hs'].rolling(7).mean()
    data["Hs_ma_30"] = data['Hs'].rolling(30).mean()
    data["Tp_ma_7"] = data['Tp'].rolling(7).mean()
    data["Tp_ma_30"] = data['Tp'].rolling(30).mean()
    data["Dir_ma_7"] = data['Dir'].rolling(7).mean()
    data["Dir_ma_30"] = data['Dir'].rolling(30).mean()
    data["Eng_ma_7"] = data['Eng'].rolling(7).mean()
    data["Eng_ma_30"] = data['Eng'].rolling(30).mean()

    data = data.dropna()

    # Split into train/test
    train = data[data['dates'].dt.year <= 2020]
    test = data[data['dates'].dt.year > 2020]

    features = ['Hs', 'Tp', 'Dir', 'Eng', 'days_since_start', 'month', 'Hs_ma_7', 'Hs_ma_30', 
                'Tp_ma_7', 'Tp_ma_30', 'Dir_ma_7', 'Dir_ma_30', 'Eng_ma_7', 'Eng_ma_30']
    target = 'shore'

    scaler = StandardScaler()
    X_train = scaler.fit_transform(train[features])
    X_test = scaler.transform(test[features])
    y_train = train[target].values
    y_test = test[target].values
    train_dates = train['dates'].values
    test_dates = test['dates'].values

    return X_train, X_test, y_train, y_test, train_dates, test_dates, scaler

# 2. Create Sequences
def create_sequences(X, y, dates, window_size=30):
    X_seq, y_seq, date_seq = [], [], []
    for i in range(len(X) - window_size):
        X_seq.append(X[i:i+window_size])
        y_seq.append(y[i+window_size])
        date_seq.append(dates[i+window_size])
    return np.array(X_seq), np.array(y_seq), np.array(date_seq)

# 3. PyTorch Model
class TCNModel(nn.Module):
    def __init__(self, input_dim, num_filters=64, kernel_size=3, dropout=0.2):
        super(TCNModel, self).__init__()

        self.pad1 = nn.ConstantPad1d((kernel_size-1, 0), 0)  # Causal padding
        self.conv1 = nn.Conv1d(input_dim, num_filters, kernel_size)
        self.bn1 = nn.BatchNorm1d(num_filters)
        self.relu1 = nn.ReLU()
        self.dropout1 = nn.Dropout(dropout)

        self.pad2 = nn.ConstantPad1d((2*(kernel_size-1), 0), 0)  # dilation=2
        self.conv2 = nn.Conv1d(num_filters, num_filters, kernel_size, dilation=2)
        self.bn2 = nn.BatchNorm1d(num_filters)
        self.relu2 = nn.ReLU()
        self.dropout2 = nn.Dropout(dropout)

        self.pad3 = nn.ConstantPad1d((4*(kernel_size-1), 0), 0)  # dilation=4
        self.conv3 = nn.Conv1d(num_filters, num_filters, kernel_size, dilation=4)
        self.bn3 = nn.BatchNorm1d(num_filters)
        self.relu3 = nn.ReLU()
        self.dropout3 = nn.Dropout(dropout)

        self.flatten = nn.Flatten()
        self.fc = nn.Linear(num_filters * 30, 1)  # 30 is the window size

    def forward(self, x):
        # x: [batch_size, seq_len, features] -> we need [batch, channels, seq_len]
        x = x.permute(0, 2, 1)

        x = self.pad1(x)
        x = self.conv1(x)
        x = self.bn1(x)
        x = self.relu1(x)
        x = self.dropout1(x)

        x = self.pad2(x)
        x = self.conv2(x)
        x = self.bn2(x)
        x = self.relu2(x)
        x = self.dropout2(x)

        x = self.pad3(x)
        x = self.conv3(x)
        x = self.bn3(x)
        x = self.relu3(x)
        x = self.dropout3(x)

        x = self.flatten(x)
        out = self.fc(x)
        return out.squeeze()


# 4. Main Execution
def main():
    # Parameters
    filepath = 'C:/Users/amgh628/Desktop/DeepLearning/Harley_waveANDshore.csv'
    WINDOW_SIZE = 30
    EPOCHS = 400
    BATCH_SIZE = 32
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')



    # Load data
    X_train, X_test, y_train, y_test, train_dates, test_dates, scaler = load_and_preprocess_data(filepath)
    
    # Create sequences
    X_train_seq, y_train_seq, train_date_seq = create_sequences(X_train, y_train, train_dates, WINDOW_SIZE)
    X_test_seq, y_test_seq, test_date_seq = create_sequences(X_test, y_test, test_dates, WINDOW_SIZE)

    # Convert to torch tensors
    X_train_seq = torch.tensor(X_train_seq, dtype=torch.float32)
    y_train_seq = torch.tensor(y_train_seq, dtype=torch.float32)
    X_test_seq = torch.tensor(X_test_seq, dtype=torch.float32)
    y_test_seq = torch.tensor(y_test_seq, dtype=torch.float32)

    # Data loaders
    train_dataset = TensorDataset(X_train_seq, y_train_seq)
    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)

    # Initialize model
    model = TCNModel(input_dim=X_train_seq.shape[2]).to(device)
    model = torch.compile(model, backend="eager")  # JIT-compile the model for better performance
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    # Early stopping parameters
    best_val_loss = np.inf
    patience = 40
    wait = 0

    # Training loop
    for epoch in range(EPOCHS):
        model.train()
        running_loss = 0.0
        for X_batch, y_batch in train_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)

            optimizer.zero_grad()
            outputs = model(X_batch)
            loss = criterion(outputs, y_batch)
            loss.backward()
            optimizer.step()
            running_loss += loss.item() * X_batch.size(0)

        epoch_loss = running_loss / len(train_loader.dataset)

        # Validation loss (simple train loss because we don't have validation split here)
        model.eval()
        with torch.no_grad():
            val_outputs = model(X_train_seq.to(device))
            val_loss = criterion(val_outputs, y_train_seq.to(device)).item()

        print(f"Epoch [{epoch+1}/{EPOCHS}], Train Loss: {epoch_loss:.4f}, Val Loss: {val_loss:.4f}")

        # Early Stopping
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_model_state = model.state_dict()
            wait = 0
        else:
            wait += 1
            if wait >= patience:
                print("Early stopping triggered.")
                break

    # Load best model
    model.load_state_dict(best_model_state)

    # Predictions
    model.eval()
    train_pred = model(X_train_seq.to(device)).cpu().detach().numpy()
    test_pred = model(X_test_seq.to(device)).cpu().detach().numpy()

    # Metrics
    print("\nEvaluation Metrics (Test):")
    print(f"Test MAE: {mean_absolute_error(y_test_seq, test_pred):.2f}")
    print(f"Test RMSE: {np.sqrt(mean_squared_error(y_test_seq, test_pred)):.2f}")
    print(f"Test R: {r2_score(y_test_seq, test_pred):.2f}")

    # Save results
    results = pd.DataFrame({
        'Date': np.concatenate([train_date_seq, test_date_seq]),
        'Actual': np.concatenate([y_train_seq.numpy(), y_test_seq.numpy()]),
        'Predicted': np.concatenate([train_pred.flatten(), test_pred.flatten()])
    })
    results.to_csv('results_pytorch_compiled.csv', index=False)

    # Plot results
    plt.figure(figsize=(14, 6))
    plt.plot(train_date_seq, y_train_seq.numpy(), label='Actual (Train)', color='blue', alpha=0.7)
    plt.plot(train_date_seq, train_pred, label='Predicted (Train)', color='cyan', linestyle='--')
    plt.plot(test_date_seq, y_test_seq.numpy(), label='Actual (Test)', color='green', alpha=0.7)
    plt.plot(test_date_seq, test_pred, label='Predicted (Test)', color='red', linestyle='--')

    plt.title('Shoreline Position Prediction (1999-2025)')
    plt.xlabel('Year')
    plt.ylabel('Shoreline Position')
    plt.legend()
    plt.grid(True)

    plt.gca().xaxis.set_major_locator(mdates.YearLocator(4))
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    plt.gcf().autofmt_xdate()

    plt.savefig('shoreline_prediction_pytorch_compiled.png', dpi=300, bbox_inches='tight')
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Set random seed for reproducibility
torch.manual_seed(42)
np.random.seed(42)

# Set device (GPU or CPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Step 1: Load and Preprocess Data
def load_data(file_path):
    """Load CSV file and parse dates."""
    df = pd.read_csv(file_path, parse_dates=['dates'], dayfirst = True)
    df.sort_values('dates', inplace=True)
    return df

# Step 2: Prepare Input Features
def prepare_features(df, window_size=60):
    """
    Create input features: lagged shoreline positions and wave parameters.
    window_size: Number of past days to use as input (default 60 for ~2 months).
    """
    features = ['shore', 'Hs', 'Tp', 'Dir', 'Eng']
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(df[features])
    scaled_df = pd.DataFrame(scaled_data, columns=features, index=df.index)
    
    # Create lagged features
    X, y = [], []
    for i in range(window_size, len(df)):
        X.append(scaled_df.iloc[i-window_size:i][features].values)  # Window of features
        y.append(scaled_df.iloc[i]['shore'])  # Next shoreline position
    X = np.array(X)  # Shape: (samples, window_size, n_features)
    y = np.array(y)  # Shape: (samples,)
    
    return X, y, scaler

# Step 3: Split Data into Training and Testing
def split_data(X, y, df, train_end_year=2021):
    """Split data into training (1999-2021) and testing (2022-2025)."""
    train_mask = df['dates'].dt.year <= train_end_year
    test_mask = df['dates'].dt.year > train_end_year
    X_train = X[train_mask[window_size:]]
    y_train = y[train_mask[window_size:]]
    X_test = X[test_mask[window_size:]]
    y_test = y[test_mask[window_size:]]
    return X_train, y_train, X_test, y_test

# Step 4: Define Custom Dataset
class TimeSeriesDataset(Dataset):
    """Custom PyTorch Dataset for time series data."""
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)
    
    def __len__(self):
        return len(self.X)
    
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

# Step 5: Define S-Transformer Model
class STransformer(nn.Module):
    """S-Transformer model for shoreline position prediction."""
    def __init__(self, window_size, n_features, d_model=64, n_heads=4, ff_dim=256, n_blocks=2, dropout=0.1):
        super(STransformer, self).__init__()
        self.window_size = window_size
        self.n_features = n_features
        self.d_model = d_model
        
        # Input Projection Layer
        self.input_projection = nn.Linear(n_features, d_model)
        
        # Positional Encoding
        self.pos_encoding = self.create_positional_encoding(window_size, d_model)
        
        # Transformer Encoder Layers
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=d_model,
            nhead=n_heads,
            dim_feedforward=ff_dim,
            dropout=dropout,
            activation='relu',
            batch_first=True
        )
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=n_blocks)
        
        # Output Layer
        self.global_pool = nn.AdaptiveAvgPool1d(1)
        self.fc = nn.Linear(d_model, 1)
    
    def create_positional_encoding(self, length, depth):
        """Create positional encoding matrix."""
        positions = np.arange(length)[:, np.newaxis]
        depths = np.arange(depth)[np.newaxis, :] / depth
        angle_rates = 1 / (10000 ** depths)
        angle_rads = positions * angle_rates
        pos_encoding = np.concatenate([np.sin(angle_rads), np.cos(angle_rads)], axis=-1)
        return torch.tensor(pos_encoding, dtype=torch.float32).to(device)
    
    def forward(self, x):
        # Project input features to d_model
        x = self.input_projection(x)  # (batch, seq_len, n_features) -> (batch, seq_len, d_model)
        
        # Add positional encoding
        x = x + self.pos_encoding[:x.size(1), :x.size(2)]
        
        # Transformer Encoder
        x = self.transformer_encoder(x)
        
        # Global Average Pooling
        x = x.permute(0, 2, 1)  # (batch, seq_len, d_model) -> (batch, d_model, seq_len)
        x = self.global_pool(x).squeeze(-1)  # (batch, d_model)
        
        # Output
        x = self.fc(x)
        return x

# Step 6: Train the Model
def train_model(model, train_loader, val_loader, epochs=200):
    """Train the S-Transformer model."""
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=1e-4)
    
    train_losses = []
    val_losses = []
    
    for epoch in range(epochs):
        model.train()
        train_loss = 0.0
        for X_batch, y_batch in train_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            optimizer.zero_grad()
            outputs = model(X_batch)
            loss = criterion(outputs.squeeze(), y_batch)
            loss.backward()
            optimizer.step()
            train_loss += loss.item() * X_batch.size(0)
        train_loss /= len(train_loader.dataset)
        train_losses.append(train_loss)
        
        # Validation
        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for X_batch, y_batch in val_loader:
                X_batch, y_batch = X_batch.to(device), y_batch.to(device)
                outputs = model(X_batch)
                loss = criterion(outputs.squeeze(), y_batch)
                val_loss += loss.item() * X_batch.size(0)
        val_loss /= len(val_loader.dataset)
        val_losses.append(val_loss)
        
        if (epoch + 1) % 10 == 0:
            print(f"Epoch {epoch+1}/{epochs}, Train Loss: {train_loss:.6f}, Val Loss: {val_loss:.6f}")
    
    return train_losses, val_losses

# Step 7: Evaluate and Predict
def evaluate_model(model, test_loader, scaler, df, window_size):
    """Evaluate model and make predictions."""
    model.eval()
    y_pred = []
    y_true = []
    
    with torch.no_grad():
        for X_batch, y_batch in test_loader:
            X_batch = X_batch.to(device)
            outputs = model(X_batch)
            y_pred.append(outputs.cpu().numpy())
            y_true.append(y_batch.numpy())
    
    y_pred = np.concatenate(y_pred).flatten()
    y_true = np.concatenate(y_true).flatten()
    
    # Inverse transform predictions and true values
    y_true = scaler.inverse_transform(np.concatenate([y_true.reshape(-1, 1), np.zeros((len(y_true), 4))], axis=1))[:, 0]
    y_pred = scaler.inverse_transform(np.concatenate([y_pred.reshape(-1, 1), np.zeros((len(y_pred), 4))], axis=1))[:, 0]
    
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    print(f"RMSE: {rmse:.4f}, R: {r2:.4f}")
    
    return y_pred, y_true

# Step 8: Plot Results
def plot_results(df, y_true, y_pred, window_size):
    """Plot observed vs. predicted shoreline positions, subsampling for daily data."""
    test_dates = df['dates'][window_size:][df['dates'].dt.year > 2021][::7]  # Subsample every 7th day
    y_true = y_true[::7]
    y_pred = y_pred[::7]
    plt.figure(figsize=(12, 6))
    plt.plot(test_dates, y_true, label='Observed Shoreline', color='blue')
    plt.plot(test_dates, y_pred, label='Predicted Shoreline', color='red', linestyle='--')
    plt.xlabel('Date')
    plt.ylabel('Shoreline Position (m)')
    plt.title('S-Transformer Shoreline Position Prediction (Daily, PyTorch)')
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Main Execution
if __name__ == "__main__":
    # Load data
    file_path = 'C:/Users/amgh628/Desktop/DeepLearning/Harley_waveANDshore.csv'  # Update with your CSV file path
    df = load_data(file_path)
    
    # Prepare features
    window_size = 60  # Number of past days (~2 months)
    X, y, scaler = prepare_features(df, window_size)
    
    # Split data
    X_train, y_train, X_test, y_test = split_data(X, y, df)
    
    # Create datasets and loaders
    train_dataset = TimeSeriesDataset(X_train, y_train)
    test_dataset = TimeSeriesDataset(X_test, y_test)
    
    # Split train into train and validation
    train_size = int(0.8 * len(train_dataset))
    val_size = len(train_dataset) - train_size
    train_subset, val_subset = torch.utils.data.random_split(train_dataset, [train_size, val_size])
    
    train_loader = DataLoader(train_subset, batch_size=32, shuffle=True)
    val_loader = DataLoader(val_subset, batch_size=32, shuffle=False)
    test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
    
    # Build and train model
    n_features = X.shape[2]  # Number of features (shoreline + 4 wave parameters)
    model = STransformer(window_size, n_features, d_model=64, n_heads=4).to(device)
    train_losses, val_losses = train_model(model, train_loader, val_loader)
    
    # Evaluate and predict
    y_pred, y_true = evaluate_model(model, test_loader, scaler, df, window_size)
    
    # Plot results
    plot_results(df, y_true, y_pred, window_size)

def plot_results(df, y_true, y_pred, window_size):
    """Plot observed vs. predicted shoreline positions, subsampling for daily data."""
    test_dates = df['dates'][window_size:][df['dates'].dt.year > 2021][::7].to_numpy()  # Convert to numpy array
    y_true = y_true[::7]
    y_pred = y_pred[::7]
    plt.figure(figsize=(12, 6))
    plt.plot(test_dates, y_true, label='Observed Shoreline', color='blue')
    plt.plot(test_dates, y_pred, label='Predicted Shoreline', color='red', linestyle='--')
    plt.xlabel('Date')
    plt.ylabel('Shoreline Position (m)')
    plt.title('S-Transformer Shoreline Position Prediction (Daily, PyTorch)')
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    # Save figure
    plt.savefig('shoreline_prediction_ST.png', dpi=300, bbox_inches='tight')
    plt.close()

# Save predictions to CSV
results = pd.DataFrame({'date': df['dates'][window_size:][df['dates'].dt.year > 2021][::7], 'observed': y_true[::7], 'predicted': y_pred[::7]})
results.to_csv('shoreline_predictions_ST.csv', index=False)

# Save figure
plt.savefig('shoreline_prediction_ST.png', dpi=300, bbox_inches='tight')
plt.close()
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torch.cuda.amp import GradScaler, autocast
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Set random seed for reproducibility
torch.manual_seed(42)
np.random.seed(42)

# Set device (GPU or CPU) and verify GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
if torch.cuda.is_available():
    print(f"GPU Name: {torch.cuda.get_device_name(0)}")
    print(f"Number of GPUs: {torch.cuda.device_count()}")
else:
    print("No GPU detected. Running on CPU.")

# Step 1: Load and Preprocess Data
def load_data(file_path):
    """Load CSV file and parse dates."""
    df = pd.read_csv(file_path, parse_dates=['date'])
    df.sort_values('date', inplace=True)
    df.reset_index(drop=True, inplace=True)  # Reset index to ensure alignment
    return df

# Step 2: Prepare Input Features
def prepare_features(df, window_size=60):
    """
    Create input features: lagged shoreline positions and wave parameters.
    Scaling will be done after splitting to avoid leakage.
    """
    features = ['shore', 'Hs', 'Tp', 'Dir', 'Eng']
    X, y = [], []
    for i in range(window_size, len(df)):
        X.append(df.iloc[i-window_size:i][features].values)  # Window of features
        y.append(df.iloc[i]['shore'])  # Next shoreline position
    X = np.array(X)  # Shape: (samples, window_size, n_features)
    y = np.array(y)  # Shape: (samples,)
    return X, y

# Step 3: Split Data into Training and Testing
def split_data(df, X, y, window_size, train_end_year=2021):
    """Split data into training (1999-2021) and testing (2022-2025), then scale features."""
    train_mask = df['date'].dt.year <= train_end_year
    test_mask = df['date'].dt.year > train_end_year
    X_train = X[train_mask[window_size:]]
    y_train = y[train_mask[window_size:]]
    X_test = X[test_mask[window_size:]]
    y_test = y[test_mask[window_size:]]
    
    # Scale features AFTER splitting to prevent data leakage
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train.reshape(-1, X_train.shape[-1])).reshape(X_train.shape)
    X_test_scaled = scaler.transform(X_test.reshape(-1, X_test.shape[-1])).reshape(X_test.shape)
    
    return X_train_scaled, y_train, X_test_scaled, y_test, scaler

# Step 4: Define Custom Dataset
class TimeSeriesDataset(Dataset):
    """Custom PyTorch Dataset for time series data."""
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)
    
    def __len__(self):
        return len(self.X)
    
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

# Step 5: Define S-Transformer Model
class STransformer(nn.Module):
    """S-Transformer model for shoreline position prediction."""
    def __init__(self, window_size, n_features, d_model=64, n_heads=4, ff_dim=256, n_blocks=2, dropout=0.2):
        super(STransformer, self).__init__()
        self.window_size = window_size
        self.n_features = n_features
        self.d_model = d_model
        
        # Input Projection Layer
        self.input_projection = nn.Linear(n_features, d_model)
        
        # Positional Encoding
        self.pos_encoding = self.create_positional_encoding(window_size, d_model)
        
        # Transformer Encoder Layers
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=d_model,
            nhead=n_heads,
            dim_feedforward=ff_dim,
            dropout=dropout,
            activation='relu',
            batch_first=True
        )
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=n_blocks)
        
        # Output Layer
        self.global_pool = nn.AdaptiveAvgPool1d(1)
        self.fc = nn.Linear(d_model, 1)
    
    def create_positional_encoding(self, length, depth):
        """Create positional encoding matrix."""
        positions = np.arange(length)[:, np.newaxis]
        depths = np.arange(depth)[np.newaxis, :] / depth
        angle_rates = 1 / (10000 ** depths)
        angle_rads = positions * angle_rates
        pos_encoding = np.concatenate([np.sin(angle_rads), np.cos(angle_rads)], axis=-1)
        return torch.tensor(pos_encoding, dtype=torch.float32).to(device)
    
    def forward(self, x):
        # Project input features to d_model
        x = self.input_projection(x)  # (batch, seq_len, n_features) -> (batch, seq_len, d_model)
        
        # Add positional encoding
        x = x + self.pos_encoding[:x.size(1), :x.size(2)]
        
        # Transformer Encoder
        x = self.transformer_encoder(x)
        
        # Global Average Pooling
        x = x.permute(0, 2, 1)  # (batch, seq_len, d_model) -> (batch, d_model, seq_len)
        x = self.global_pool(x).squeeze(-1)  # (batch, d_model)
        
        # Output
        x = self.fc(x)
        return x

# Step 6: Train the Model
def train_model(model, train_loader, val_loader, epochs=200):
    """Train the S-Transformer model with mixed precision."""
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=1e-4)
    scaler = GradScaler()
    
    train_losses = []
    val_losses = []
    
    for epoch in range(epochs):
        model.train()
        train_loss = 0.0
        for i, (X_batch, y_batch) in enumerate(train_loader):
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            if epoch == 0 and i == 0:
                print(f"X_batch device: {X_batch.device}, y_batch device: {y_batch.device}")
            optimizer.zero_grad()
            with autocast():
                outputs = model(X_batch)
                loss = criterion(outputs.squeeze(), y_batch)
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
            train_loss += loss.item() * X_batch.size(0)
        train_loss /= len(train_loader.dataset)
        train_losses.append(train_loss)
        
        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for X_batch, y_batch in val_loader:
                X_batch, y_batch = X_batch.to(device), y_batch.to(device)
                with autocast():
                    outputs = model(X_batch)
                    loss = criterion(outputs.squeeze(), y_batch)
                val_loss += loss.item() * X_batch.size(0)
        val_loss /= len(val_loader.dataset)
        val_losses.append(val_loss)
        
        if (epoch + 1) % 10 == 0:
            print(f"Epoch {epoch+1}/{epochs}, Train Loss: {train_loss:.6f}, Val Loss: {val_loss:.6f}")
    
    return train_losses, val_losses

# Step 7: Evaluate and Predict
def evaluate_model(model, test_loader):
    """Evaluate model and make predictions."""
    model.eval()
    y_pred = []
    y_true = []
    
    with torch.no_grad():
        for X_batch, y_batch in test_loader:
            X_batch = X_batch.to(device)
            with autocast():
                outputs = model(X_batch)
            y_pred.append(outputs.cpu().numpy())
            y_true.append(y_batch.numpy())
    
    y_pred = np.concatenate(y_pred).flatten()
    y_true = np.concatenate(y_true).flatten()
    
    # No scaling applied to y, so no inverse transform needed
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    print(f"RMSE: {rmse:.4f}, R: {r2:.4f}")
    
    return y_pred, y_true

# Step 8: Plot Results
def plot_results(df, y_true_train, y_pred_train, y_true_test, y_pred_test, window_size):
    """Plot observed vs. predicted shoreline positions for training and testing periods with date and time on x-axis."""
    # Training period (1999-2021)
    train_dates = df['date'][window_size:][df['date'].dt.year <= 2021][::7].to_numpy()
    y_true_train = y_true_train[::7]
    y_pred_train = y_pred_train[::7]
    plt.figure(figsize=(12, 6))
    plt.plot(train_dates, y_true_train, label='Observed Shoreline', color='blue')
    plt.plot(train_dates, y_pred_train, label='Predicted Shoreline', color='red', linestyle='--')
    plt.xlabel('Date and Time')
    plt.ylabel('Shoreline Position (m)')
    plt.title('S-Transformer Shoreline Position Prediction (Training, 1999-2021)')
    plt.legend()
    plt.xticks(rotation=45)
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))
    plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())
    plt.tight_layout()
    plt.savefig('shoreline_prediction_train.png', dpi=300, bbox_inches='tight')
    plt.close()

    # Testing period (2022-2025)
    test_dates = df['date'][window_size:][df['date'].dt.year > 2021][::7].to_numpy()
    y_true_test = y_true_test[::7]
    y_pred_test = y_pred_test[::7]
    plt.figure(figsize=(12, 6))
    plt.plot(test_dates, y_true_test, label='Observed Shoreline', color='blue')
    plt.plot(test_dates, y_pred_test, label='Predicted Shoreline', color='red', linestyle='--')
    plt.xlabel('Date and Time')
    plt.ylabel('Shoreline Position (m)')
    plt.title('S-Transformer Shoreline Position Prediction (Testing, 2022-2025)')
    plt.legend()
    plt.xticks(rotation=45)
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))
    plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())
    plt.tight_layout()
    plt.savefig('shoreline_prediction_test.png', dpi=300, bbox_inches='tight')
    plt.close()

# Main Execution
if __name__ == "__main__":
    # Load data
    file_path = 'C:/Users/amgh628/Desktop/DeepLearning/Harley_waveANDshore.csv'  # Update with your CSV file path
    df = load_data(file_path)
    
    # Prepare features without scaling
    window_size = 60
    X, y = prepare_features(df, window_size)
    
    # Split and scale data
    X_train, y_train, X_test, y_test, scaler = split_data(df, X, y, window_size)
    
    # Create datasets and loaders
    train_dataset = TimeSeriesDataset(X_train, y_train)
    test_dataset = TimeSeriesDataset(X_test, y_test)
    
    # Split train into train and validation
    train_size = int(0.8 * len(train_dataset))
    val_size = len(train_dataset) - train_size
    train_subset, val_subset = torch.utils.data.random_split(train_dataset, [train_size, val_size])
    
    train_loader = DataLoader(train_subset, batch_size=32, shuffle=True)
    val_loader = DataLoader(val_subset, batch_size=32, shuffle=False)
    test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
    
    # Build and train model
    n_features = X.shape[2]
    model = STransformer(window_size, n_features, d_model=64, n_heads=4).to(device)
    train_losses, val_losses = train_model(model, train_loader, val_loader)
    
    # Evaluate and predict for both periods
    full_dataset = TimeSeriesDataset(X, y)
    full_loader = DataLoader(full_dataset, batch_size=32, shuffle=False)
    y_pred_full, y_true_full = evaluate_model(model, full_loader)
    y_pred_train = y_pred_full[df['date'][window_size:].dt.year <= 2021]
    y_true_train = y_true_full[df['date'][window_size:].dt.year <= 2021]
    y_pred_test = y_pred_full[df['date'][window_size:].dt.year > 2021]
    y_true_test = y_true_full[df['date'][window_size:].dt.year > 2021]
    
    # Plot results
    plot_results(df, y_true_train, y_pred_train, y_true_test, y_pred_test, window_size)
    
    # Save predictions to CSV
    results = pd.DataFrame({
        'date': df['date'][window_size:][::7],
        'observed': y_true_full[::7],
        'predicted': y_pred_full[::7]
    })
    results.to_csv('shoreline_predictions.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import TimeSeriesSplit
import matplotlib.pyplot as plt
import os

# Set random seed for reproducibility
torch.manual_seed(42)
np.random.seed(42)

# Set device (GPU or CPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Step 1: Load and Preprocess Data
def load_data(file_path):
    """Load CSV file and parse dates."""
    df = pd.read_csv(file_path, parse_dates=['dates'], dayfirst=True)
    df.sort_values('dates', inplace=True)
    return df

# Step 2: Prepare Input Features
def prepare_features(df, window_size=60):
    """
    Create input features: lagged shoreline positions and wave parameters.
    Fit scaler only on training data to prevent leakage.
    """
    features = ['shore', 'Hs', 'Tp', 'Dir', 'Eng']
    
    # Split data into train and test based on year
    train_mask = df['dates'].dt.year <= 2021
    train_df = df[train_mask]
    
    # Fit scaler on training data only
    scaler = StandardScaler()
    scaler.fit(train_df[features])
    
    # Transform all data
    scaled_data = scaler.transform(df[features])
    scaled_df = pd.DataFrame(scaled_data, columns=features, index=df.index)
    
    # Create lagged features
    X, y = [], []
    for i in range(window_size, len(df)):
        X.append(scaled_df.iloc[i-window_size:i][features].values)
        y.append(scaled_df.iloc[i]['shore'])
    X = np.array(X)  # Shape: (samples, window_size, n_features)
    y = np.array(y)  # Shape: (samples,)
    
    return X, y, scaler

# Step 3: Split Data into Training and Testing
def split_data(X, y, df, train_end_year=2021):
    """Split data into training (1999-2021) and testing (2022-2025)."""
    train_mask = df['dates'].dt.year <= train_end_year
    test_mask = df['dates'].dt.year > train_end_year
    X_train = X[train_mask[window_size:]]
    y_train = y[train_mask[window_size:]]
    X_test = X[test_mask[window_size:]]
    y_test = y[test_mask[window_size:]]
    return X_train, y_train, X_test, y_test

# Step 4: Define Custom Dataset
class TimeSeriesDataset(Dataset):
    """Custom PyTorch Dataset for time series data."""
    def __init__(self, X, y, augment=False):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)
        self.augment = augment
    
    def __len__(self):
        return len(self.X)
    
    def __getitem__(self, idx):
        X, y = self.X[idx], self.y[idx]
        if self.augment:
            # Add subtle noise to input features
            noise = torch.normal(0, 0.01, size=X.shape).to(X.device)
            X = X + noise
        return X, y

# Step 5: Define S-Transformer Model
class STransformer(nn.Module):
    """S-Transformer model for shoreline position prediction."""
    def __init__(self, window_size, n_features, d_model=32, n_heads=4, ff_dim=128, n_blocks=1, dropout=0.3):
        super(STransformer, self).__init__()
        self.window_size = window_size
        self.n_features = n_features
        self.d_model = d_model
        
        # Input Projection Layer
        self.input_projection = nn.Linear(n_features, d_model)
        self.norm = nn.LayerNorm(d_model)
        
        # Positional Encoding
        self.pos_encoding = self.create_positional_encoding(window_size, d_model)
        
        # Transformer Encoder Layers
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=d_model,
            nhead=n_heads,
            dim_feedforward=ff_dim,
            dropout=dropout,
            activation='relu',
            batch_first=True
        )
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=n_blocks)
        
        # Output Layer
        self.global_pool = nn.AdaptiveAvgPool1d(1)
        self.fc = nn.Linear(d_model, 1)
    
    def create_positional_encoding(self, length, depth):
        """Create positional encoding matrix."""
        positions = np.arange(length)[:, np.newaxis]
        depths = np.arange(depth)[np.newaxis, :] / depth
        angle_rates = 1 / (10000 ** depths)
        angle_rads = positions * angle_rates
        pos_encoding = np.concatenate([np.sin(angle_rads), np.cos(angle_rads)], axis=-1)
        return torch.tensor(pos_encoding, dtype=torch.float32).to(device)
    
    def forward(self, x):
        # Project input features to d_model
        x = self.input_projection(x)
        x = self.norm(x)
        
        # Add positional encoding
        x = x + self.pos_encoding[:x.size(1), :x.size(2)]
        
        # Transformer Encoder
        x = self.transformer_encoder(x)
        
        # Global Average Pooling
        x = x.permute(0, 2, 1)
        x = self.global_pool(x).squeeze(-1)
        
        # Output
        x = self.fc(x)
        return x

# Step 6: Train the Model
def train_model(model, train_loader, val_loader, epochs=200, patience=20):
    """Train the S-Transformer model with early stopping."""
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=1e-4, weight_decay=1e-5)
    
    train_losses = []
    val_losses = []
    best_val_loss = float('inf')
    patience_counter = 0
    
    for epoch in range(epochs):
        model.train()
        train_loss = 0.0
        for X_batch, y_batch in train_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            optimizer.zero_grad()
            outputs = model(X_batch)
            loss = criterion(outputs.squeeze(), y_batch)
            loss.backward()
            optimizer.step()
            train_loss += loss.item() * X_batch.size(0)
        train_loss /= len(train_loader.dataset)
        train_losses.append(train_loss)
        
        # Validation
        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for X_batch, y_batch in val_loader:
                X_batch, y_batch = X_batch.to(device), y_batch.to(device)
                outputs = model(X_batch)
                loss = criterion(outputs.squeeze(), y_batch)
                val_loss += loss.item() * X_batch.size(0)
        val_loss /= len(val_loader.dataset)
        val_losses.append(val_loss)
        
        # Early stopping
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            patience_counter = 0
            torch.save(model.state_dict(), 'best_model.pt')
        else:
            patience_counter += 1
            if patience_counter >= patience:
                print(f"Early stopping at epoch {epoch+1}")
                break
        
        if (epoch + 1) % 10 == 0:
            print(f"Epoch {epoch+1}/{epochs}, Train Loss: {train_loss:.6f}, Val Loss: {val_loss:.6f}")
    
    # Load best model
    if os.path.exists('best_model.pt'):
        model.load_state_dict(torch.load('best_model.pt'))
    return train_losses, val_losses

# Step 7: Evaluate and Predict
def evaluate_model(model, test_loader, scaler, df, window_size):
    """Evaluate model and make predictions."""
    model.eval()
    y_pred = []
    y_true = []
    
    with torch.no_grad():
        for X_batch, y_batch in test_loader:
            X_batch = X_batch.to(device)
            outputs = model(X_batch)
            y_pred.append(outputs.cpu().numpy())
            y_true.append(y_batch.numpy())
    
    y_pred = np.concatenate(y_pred).flatten()
    y_true = np.concatenate(y_true).flatten()
    
    # Inverse transform predictions and true values
    y_true = scaler.inverse_transform(np.concatenate([y_true.reshape(-1, 1), np.zeros((len(y_true), 4))], axis=1))[:, 0]
    y_pred = scaler.inverse_transform(np.concatenate([y_pred.reshape(-1, 1), np.zeros((len(y_pred), 4))], axis=1))[:, 0]
    
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    print(f"Test RMSE: {rmse:.4f}, Test R: {r2:.4f}")
    
    return y_pred, y_true

# Step 8: Plot Results
def plot_results(df, y_true, y_pred, window_size, train_losses, val_losses):
    """Plot observed vs. predicted shoreline positions and loss curves."""
    # Subsample for plotting
    test_dates = df['dates'][window_size:][df['dates'].dt.year > 2021][::7].to_numpy()
    y_true_sub = y_true[::7]
    y_pred_sub = y_pred[::7]
    
    # Plot predictions
    plt.figure(figsize=(12, 6))
    plt.plot(test_dates, y_true_sub, label='Observed Shoreline', color='blue')
    plt.plot(test_dates, y_pred_sub, label='Predicted Shoreline', color='red', linestyle='--')
    plt.xlabel('Date')
    plt.ylabel('Shoreline Position (m)')
    plt.title('S-Transformer Shoreline Position Prediction (Daily, PyTorch)')
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig('shoreline_prediction_ST.png', dpi=300, bbox_inches='tight')
    plt.close()
    
    # Plot training and validation losses
    plt.figure(figsize=(10, 5))
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss')
    plt.legend()
    plt.savefig('loss_curves_ST.png', dpi=300, bbox_inches='tight')
    plt.close()

# Step 9: Optional Cross-Validation
def cross_validate_model(X_train, y_train, window_size, n_features, n_splits=5):
    """Perform time-series cross-validation."""
    tscv = TimeSeriesSplit(n_splits=n_splits)
    rmses, r2s = [], []
    
    for train_idx, val_idx in tscv.split(X_train):
        X_tr, X_val = X_train[train_idx], X_train[val_idx]
        y_tr, y_val = y_train[train_idx], y_train[val_idx]
        
        train_dataset = TimeSeriesDataset(X_tr, y_tr, augment=True)
        val_dataset = TimeSeriesDataset(X_val, y_val)
        train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
        val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
        
        model = STransformer(window_size, n_features, d_model=32, n_heads=4, ff_dim=128, n_blocks=1).to(device)
        train_model(model, train_loader, val_loader, epochs=200, patience=20)
        
        model.eval()
        y_pred = []
        y_true = []
        with torch.no_grad():
            for X_batch, y_batch in val_loader:
                X_batch = X_batch.to(device)
                outputs = model(X_batch)
                y_pred.append(outputs.cpu().numpy())
                y_true.append(y_batch.numpy())
        
        y_pred = np.concatenate(y_pred).flatten()
        y_true = np.concatenate(y_true).flatten()
        rmse = np.sqrt(mean_squared_error(y_true, y_pred))
        r2 = r2_score(y_true, y_pred)
        rmses.append(rmse)
        r2s.append(r2)
    
    print(f"Cross-Validation RMSE: {np.mean(rmses):.4f}  {np.std(rmses):.4f}")
    print(f"Cross-Validation R: {np.mean(r2s):.4f}  {np.std(r2s):.4f}")

# Main Execution
if __name__ == "__main__":
    # Load data
    file_path = 'C:/Users/amgh628/Desktop/DeepLearning/Harley_waveANDshore.csv'  # Update with your CSV file path
    df = load_data(file_path)
    
    # Prepare features
    window_size = 60
    X, y, scaler = prepare_features(df, window_size)
    
    # Split data
    X_train, y_train, X_test, y_test = split_data(X, y, df)
    
    # Create datasets and loaders
    train_dataset = TimeSeriesDataset(X_train, y_train, augment=True)
    test_dataset = TimeSeriesDataset(X_test, y_test)
    
    # Split train into train and validation
    train_size = int(0.8 * len(train_dataset))
    val_size = len(train_dataset) - train_size
    train_subset, val_subset = torch.utils.data.random_split(train_dataset, [train_size, val_size])
    
    train_loader = DataLoader(train_subset, batch_size=32, shuffle=True)
    val_loader = DataLoader(val_subset, batch_size=32, shuffle=False)
    test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
    
    # Build and train model
    n_features = X.shape[2]
    model = STransformer(window_size, n_features, d_model=32, n_heads=4, ff_dim=128, n_blocks=1).to(device)
    
    # Print model size
    print(f"Model parameters: {sum(p.numel() for p in model.parameters())}")
    
    train_losses, val_losses = train_model(model, train_loader, val_loader, epochs=200, patience=20)
    
    # Evaluate and predict
    y_pred, y_true = evaluate_model(model, test_loader, scaler, df, window_size)
    
    # Plot results
    plot_results(df, y_true, y_pred, window_size, train_losses, val_losses)
    
    # Save predictions to CSV
    results = pd.DataFrame({
        'date': df['dates'][window_size:][df['dates'].dt.year > 2021][::7],
        'observed': y_true[::7],
        'predicted': y_pred[::7]
    })
    results.to_csv('shoreline_predictions_ST.csv', index=False)
    
    # Optional: Cross-validation (uncomment to run)
    # cross_validate_model(X_train, y_train, window_size, n_features, n_splits=5)
    
    # Optional: Feature importance via permutation (uncomment to run)
    """
    from sklearn.inspection import permutation_importance
    def compute_permutation_importance(model, X_test, y_test, feature_names):
        model.eval()
        X_test_tensor = torch.tensor(X_test, dtype=torch.float32).to(device)
        y_pred = model(X_test_tensor).cpu().detach().numpy().flatten()
        baseline_mse = mean_squared_error(y_test, y_pred)
        
        importance = []
        for col in range(X_test.shape[2]):
            X_permuted = X_test.copy()
            np.random.shuffle(X_permuted[:, :, col])
            X_permuted_tensor = torch.tensor(X_permuted, dtype=torch.float32).to(device)
            y_pred_permuted = model(X_permuted_tensor).cpu().detach().numpy().flatten()
            mse_permuted = mean_squared_error(y_test, y_pred_permuted)
            importance.append(mse_permuted - baseline_mse)
        
        importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importance})
        print(importance_df.sort_values(by='Importance', ascending=False))
    
    compute_permutation_importance(model, X_test, y_test, ['shore', 'Hs', 'Tp', 'Dir', 'Eng'])
    """
```

```python
# Full, fixed code
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import random
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Conv1D, BatchNormalization, ReLU, Dropout, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# 0. Set random seeds for reproducibility
def set_seed(seed=42):
    np.random.seed(seed)
    random.seed(seed)
    tf.random.set_seed(seed)

set_seed(42)

# 1. Data Preparation
def load_and_preprocess_data(filepath):
    data = pd.read_csv(filepath, parse_dates=['dates'])

    # Create temporal features
    data['days_since_start'] = (data['dates'] - data['dates'].min()).dt.days
    data['month'] = data['dates'].dt.month
    data["Hs_ma_7"] = data['Hs'].rolling(7).mean()
    data["Hs_ma_30"] = data['Hs'].rolling(30).mean()
    data["Tp_ma_7"] = data['Tp'].rolling(7).mean()
    data["Tp_ma_30"] = data['Tp'].rolling(30).mean()
    data["Dir_ma_7"] = data['Dir'].rolling(7).mean()
    data["Dir_ma_30"] = data['Dir'].rolling(30).mean()
    data["Eng_ma_7"] = data['Eng'].rolling(7).mean()
    data["Eng_ma_30"] = data['Eng'].rolling(30).mean()

    data = data.dropna()

    train = data[data['dates'].dt.year <= 2020]
    test = data[data['dates'].dt.year > 2020]

    features = ['Hs', 'Tp', 'Dir', 'Eng', 'days_since_start', 'month', 'Hs_ma_7', 'Hs_ma_30', 
                'Tp_ma_7', 'Tp_ma_30', 'Dir_ma_7', 'Dir_ma_30', 'Eng_ma_7', 'Eng_ma_30']
    target = 'shore'

    scaler = StandardScaler()
    X_train = scaler.fit_transform(train[features])
    X_test = scaler.transform(test[features])
    y_train = train[target].values
    y_test = test[target].values
    train_dates = train['dates'].values
    test_dates = test['dates'].values

    return X_train, X_test, y_train, y_test, train_dates, test_dates, scaler

# 2. Create Sequences
def create_sequences(X, y, dates, window_size=30):
    X_seq, y_seq, date_seq = [], [], []
    for i in range(len(X) - window_size):
        X_seq.append(X[i:i+window_size])
        y_seq.append(y[i+window_size])
        date_seq.append(dates[i+window_size])
    return np.array(X_seq), np.array(y_seq), np.array(date_seq)

# 3. TCN Model
def build_tcn(input_shape, num_filters=64, kernel_size=3, dropout=0.3):
    inputs = Input(shape=input_shape)

    x = Conv1D(num_filters, kernel_size, padding='causal', dilation_rate=1)(inputs)
    x = BatchNormalization()(x)
    x = ReLU()(x)
    x = Dropout(dropout)(x)

    x = Conv1D(num_filters, kernel_size, padding='causal', dilation_rate=2)(x)
    x = BatchNormalization()(x)
    x = ReLU()(x)
    x = Dropout(dropout)(x)

    x = Conv1D(num_filters, kernel_size, padding='causal', dilation_rate=4)(x)
    x = BatchNormalization()(x)
    x = ReLU()(x)
    x = Dropout(dropout)(x)

    x = Flatten()(x)
    outputs = Dense(1)(x)

    return Model(inputs, outputs)

# 4. Main Execution
def main():
    WINDOW_SIZE = 30
    EPOCHS = 300
    BATCH_SIZE = 32

    X_train, X_test, y_train, y_test, train_dates, test_dates, scaler = load_and_preprocess_data(
        'C:/Users/amgh628/Desktop/DeepLearning/OTAMA_linear_smoothed.csv')

    X_train_seq, y_train_seq, train_date_seq = create_sequences(X_train, y_train, train_dates, WINDOW_SIZE)
    X_test_seq, y_test_seq, test_date_seq = create_sequences(X_test, y_test, test_dates, WINDOW_SIZE)

    model = build_tcn((WINDOW_SIZE, X_train_seq.shape[2]))
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

    early_stop = EarlyStopping(monitor='val_loss', patience=80, min_delta=0.001, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=30, verbose=1)


    history = model.fit(X_train_seq, y_train_seq,
                        epochs=EPOCHS,
                        batch_size=BATCH_SIZE,
                        validation_split=0.2,
                        callbacks=[early_stop, reduce_lr],
                        verbose=1)

    # Predict
    train_pred = model.predict(X_train_seq)
    test_pred = model.predict(X_test_seq)

    # Metrics
    print("\nEvaluation Metrics (Test):")
    print(f"Test MAE: {mean_absolute_error(y_test_seq, test_pred):.2f}")
    print(f"Test RMSE: {np.sqrt(mean_squared_error(y_test_seq, test_pred)):.2f}")
    print(f"Test R: {r2_score(y_test_seq, test_pred):.2f}")


    # Plot
    plt.figure(figsize=(14, 6))
    plt.plot(train_date_seq, y_train_seq, label='Actual (Train)', color='blue', alpha=0.7)
    plt.plot(train_date_seq, train_pred, label='Predicted (Train)', color='cyan', linestyle='--')
    plt.plot(test_date_seq, y_test_seq, label='Actual (Test)', color='green', alpha=0.7)
    plt.plot(test_date_seq, test_pred, label='Predicted (Test)', color='red', linestyle='--')

    plt.title('Shoreline Position Prediction (1999-2025)')
    plt.xlabel('Year')
    plt.ylabel('Shoreline Position')
    plt.legend()
    plt.grid(True)

    plt.gca().xaxis.set_major_locator(mdates.YearLocator(4))
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    plt.gcf().autofmt_xdate()

    plt.savefig('shoreline_prediction.png', dpi=300, bbox_inches='tight')
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torch.cuda.amp import GradScaler, autocast
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import math

# Set random seed for reproducibility
torch.manual_seed(42)
np.random.seed(42)

# Set device (GPU or CPU) and verify GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
if torch.cuda.is_available():
    print(f"GPU Name: {torch.cuda.get_device_name(0)}")
    print(f"Number of GPUs: {torch.cuda.device_count()}")
else:
    print("No GPU detected. Running on CPU.")

# Step 1: Load and Preprocess Data with consistent date column naming
def load_data(file_path):
    """Load CSV file and parse dates with consistent naming."""
    df = pd.read_csv(file_path, parse_dates=['dates'])  # Ensure this column name matches your CSV
    df.rename(columns={'dates': 'date'}, inplace=True)  # Standardize to 'date' for consistency
    df.sort_values('date', inplace=True)
    df.reset_index(drop=True, inplace=True)
    return df

# Step 2: Prepare data with proper time series split to avoid leakage
def prepare_data(df, window_size=60, train_end_year=2021):
    """
    Prepare data with proper time series split to avoid leakage:
    1. Split first by time (all training data before test data)
    2. Scale features using only training data
    3. Create windows separately for train and test sets
    """
    # Define input features and target
    input_features = ['Hs', 'Tp', 'Dir', 'Eng']  # Exclude 'shore' from inputs to avoid direct leakage
    target = 'shore'
    
    # First split the dataframe by year
    train_df = df[df['date'].dt.year <= train_end_year].copy()
    test_df = df[df['date'].dt.year > train_end_year].copy()
    
    # Scale features using only training data
    scaler = StandardScaler()
    train_df[input_features] = scaler.fit_transform(train_df[input_features])
    test_df[input_features] = scaler.transform(test_df[input_features])
    
    # Scale target separately (optional but can help training)
    target_scaler = StandardScaler()
    train_df[target] = target_scaler.fit_transform(train_df[[target]])
    test_df[target] = target_scaler.transform(test_df[[target]])
    
    # Function to create windows
    def create_windows(dataset, features_list):
        X, y, dates = [], [], []
        for i in range(window_size, len(dataset)):
            # Include previous shoreline positions in the window
            window_features = dataset.iloc[i-window_size:i][features_list + [target]].values
            X.append(window_features)
            y.append(dataset.iloc[i][target])
            dates.append(dataset.iloc[i]['date'])
        return np.array(X), np.array(y), np.array(dates)
    
    X_train, y_train, train_dates = create_windows(train_df, input_features)
    X_test, y_test, test_dates = create_windows(test_df, input_features)
    
    return X_train, y_train, train_dates, X_test, y_test, test_dates, scaler, target_scaler

# Step 3: Define Custom Dataset
class TimeSeriesDataset(Dataset):
    """Custom PyTorch Dataset for time series data."""
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)
    
    def __len__(self):
        return len(self.X)
    
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

# Step 4: Define S-Transformer Model with Fixed Positional Encoding
class STransformer(nn.Module):
    """S-Transformer model for shoreline position prediction."""
    def __init__(self, window_size, n_features, d_model=64, n_heads=4, ff_dim=256, n_blocks=2, dropout=0.2):
        super(STransformer, self).__init__()
        self.window_size = window_size
        self.n_features = n_features
        self.d_model = d_model
        
        # Input Projection Layer
        self.input_projection = nn.Linear(n_features, d_model)
        
        # Positional Encoding - fixed implementation
        self.register_buffer('pos_encoding', self.create_positional_encoding(window_size, d_model))
        
        # Transformer Encoder Layers
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=d_model,
            nhead=n_heads,
            dim_feedforward=ff_dim,
            dropout=dropout,
            activation='relu',
            batch_first=True
        )
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=n_blocks)
        
        # Output Layer
        self.global_pool = nn.AdaptiveAvgPool1d(1)
        self.fc = nn.Linear(d_model, 1)
    
    def create_positional_encoding(self, length, depth):
        """Create positional encoding matrix using PyTorch."""
        positions = torch.arange(length).unsqueeze(1).float()
        div_term = torch.exp(torch.arange(0, depth, 2).float() * (-math.log(10000.0) / depth))
        pos_encoding = torch.zeros(length, depth)
        pos_encoding[:, 0::2] = torch.sin(positions * div_term)
        pos_encoding[:, 1::2] = torch.cos(positions * div_term[:pos_encoding.size(1)//2])
        return pos_encoding
    
    def forward(self, x):
        # Project input features to d_model
        x = self.input_projection(x)  # (batch, seq_len, n_features) -> (batch, seq_len, d_model)
        
        # Add positional encoding
        x = x + self.pos_encoding[:x.size(1), :].unsqueeze(0)
        
        # Transformer Encoder
        x = self.transformer_encoder(x)
        
        # Global Average Pooling
        x = x.permute(0, 2, 1)  # (batch, seq_len, d_model) -> (batch, d_model, seq_len)
        x = self.global_pool(x).squeeze(-1)  # (batch, d_model)
        
        # Output
        x = self.fc(x)
        return x

# Step 5: Train the Model
def train_model(model, train_loader, val_loader, epochs=200, patience=20):
    """Train the S-Transformer model with mixed precision and early stopping."""
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=1e-4)
    scaler = GradScaler()
    
    train_losses = []
    val_losses = []
    
    best_val_loss = float('inf')
    patience_counter = 0
    best_model_state = None
    
    for epoch in range(epochs):
        # Training phase
        model.train()
        train_loss = 0.0
        for i, (X_batch, y_batch) in enumerate(train_loader):
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            optimizer.zero_grad()
            with autocast():
                outputs = model(X_batch)
                loss = criterion(outputs.squeeze(), y_batch)
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
            train_loss += loss.item() * X_batch.size(0)
        train_loss /= len(train_loader.dataset)
        train_losses.append(train_loss)
        
        # Validation phase
        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for X_batch, y_batch in val_loader:
                X_batch, y_batch = X_batch.to(device), y_batch.to(device)
                with autocast():
                    outputs = model(X_batch)
                    loss = criterion(outputs.squeeze(), y_batch)
                val_loss += loss.item() * X_batch.size(0)
        val_loss /= len(val_loader.dataset)
        val_losses.append(val_loss)
        
        # Print progress
        if (epoch + 1) % 10 == 0:
            print(f"Epoch {epoch+1}/{epochs}, Train Loss: {train_loss:.6f}, Val Loss: {val_loss:.6f}")
        
        # Early stopping logic
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            patience_counter = 0
            best_model_state = {key: value.cpu() for key, value in model.state_dict().items()}
        else:
            patience_counter += 1
            if patience_counter >= patience:
                print(f"Early stopping triggered after {epoch+1} epochs")
                model.load_state_dict(best_model_state)
                for param in best_model_state:
                    model.state_dict()[param].copy_(best_model_state[param])
                break
    
    return train_losses, val_losses

# Step 6: Evaluate and Predict
def evaluate_model(model, data_loader, target_scaler=None):
    """Evaluate model and make predictions, with optional inverse scaling."""
    model.eval()
    y_pred = []
    y_true = []
    
    with torch.no_grad():
        for X_batch, y_batch in data_loader:
            X_batch = X_batch.to(device)
            with autocast():
                outputs = model(X_batch)
            y_pred.append(outputs.cpu().numpy())
            y_true.append(y_batch.numpy())
    
    y_pred = np.concatenate(y_pred).flatten()
    y_true = np.concatenate(y_true).flatten()
    
    # Inverse transform if scaler was provided
    if target_scaler:
        y_pred = target_scaler.inverse_transform(y_pred.reshape(-1, 1)).flatten()
        y_true = target_scaler.inverse_transform(y_true.reshape(-1, 1)).flatten()
    
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    print(f"RMSE: {rmse:.4f}, R: {r2:.4f}")
    
    return y_pred, y_true

# Step 7: Plot Results
def plot_results(train_dates, y_true_train, y_pred_train, test_dates, y_true_test, y_pred_test, 
                 sample_interval=7):
    """Plot observed vs. predicted shoreline positions for training and testing periods"""
    # Training period
    sampled_indices_train = np.arange(0, len(train_dates), sample_interval)
    train_dates_sampled = train_dates[sampled_indices_train]
    y_true_train_sampled = y_true_train[sampled_indices_train]
    y_pred_train_sampled = y_pred_train[sampled_indices_train]
    
    plt.figure(figsize=(12, 6))
    plt.plot(train_dates_sampled, y_true_train_sampled, label='Observed Shoreline', color='blue')
    plt.plot(train_dates_sampled, y_pred_train_sampled, label='Predicted Shoreline', color='red', linestyle='--')
    plt.xlabel('Date and Time')
    plt.ylabel('Shoreline Position (m)')
    plt.title('S-Transformer Shoreline Position Prediction (Training Period)')
    plt.legend()
    plt.xticks(rotation=45)
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())
    plt.tight_layout()
    plt.savefig('shoreline_prediction_train.png', dpi=300, bbox_inches='tight')
    plt.close()

    # Testing period
    sampled_indices_test = np.arange(0, len(test_dates), sample_interval)
    test_dates_sampled = test_dates[sampled_indices_test]
    y_true_test_sampled = y_true_test[sampled_indices_test]
    y_pred_test_sampled = y_pred_test[sampled_indices_test]
    
    plt.figure(figsize=(12, 6))
    plt.plot(test_dates_sampled, y_true_test_sampled, label='Observed Shoreline', color='blue')
    plt.plot(test_dates_sampled, y_pred_test_sampled, label='Predicted Shoreline', color='red', linestyle='--')
    plt.xlabel('Date and Time')
    plt.ylabel('Shoreline Position (m)')
    plt.title('S-Transformer Shoreline Position Prediction (Testing Period)')
    plt.legend()
    plt.xticks(rotation=45)
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())
    plt.tight_layout()
    plt.savefig('shoreline_prediction_test.png', dpi=300, bbox_inches='tight')
    plt.close()
    
    # Error distribution plot
    plt.figure(figsize=(10, 6))
    train_errors = y_pred_train - y_true_train
    test_errors = y_pred_test - y_true_test
    plt.hist(train_errors, bins=30, alpha=0.7, label='Train Errors')
    plt.hist(test_errors, bins=30, alpha=0.7, label='Test Errors')
    plt.axvline(x=0, color='k', linestyle='--')
    plt.xlabel('Prediction Error (m)')
    plt.ylabel('Frequency')
    plt.title('Error Distribution: Train vs Test')
    plt.legend()
    plt.tight_layout()
    plt.savefig('error_distribution.png', dpi=300, bbox_inches='tight')
    plt.close()

# Step 8: Implement a simple baseline model for comparison
def baseline_model(X_train, y_train, X_test):
    """
    A simple baseline model that uses an average of the last N shoreline positions
    from the input window to predict the next shoreline position.
    """
    # Assumes last feature column is 'shore' in X_train and X_test
    shore_col_idx = -1  # Adjust if the shore column is not the last one
    
    # For training data, predict using mean of last 5 shoreline positions in window
    y_pred_train = np.mean(X_train[:, -5:, shore_col_idx], axis=1)
    
    # For test data, predict using the same approach
    y_pred_test = np.mean(X_test[:, -5:, shore_col_idx], axis=1)
    
    return y_pred_train, y_pred_test

# Main Execution
if __name__ == "__main__":
    # Load data
    file_path = 'C:/Users/amgh628/Desktop/DeepLearning/Harley_waveANDshore.csv'  # Update with your CSV file path
    df = load_data(file_path)
    
    # Prepare data with proper time series split
    window_size = 60
    train_end_year = 2021
    X_train, y_train, train_dates, X_test, y_test, test_dates, scaler, target_scaler = prepare_data(
        df, window_size, train_end_year)
    
    # Split train into train and validation (temporal split - last 20% of training data)
    val_split_idx = int(0.8 * len(X_train))
    X_train_final = X_train[:val_split_idx]
    y_train_final = y_train[:val_split_idx]
    X_val = X_train[val_split_idx:]
    y_val = y_train[val_split_idx:]
    
    # Create datasets and loaders
    train_dataset = TimeSeriesDataset(X_train_final, y_train_final)
    val_dataset = TimeSeriesDataset(X_val, y_val)
    test_dataset = TimeSeriesDataset(X_test, y_test)
    
    # Create full train dataset for final evaluation
    full_train_dataset = TimeSeriesDataset(X_train, y_train)
    
    # DataLoaders with appropriate settings for time series
    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=False)
    val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
    test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
    full_train_loader = DataLoader(full_train_dataset, batch_size=32, shuffle=False)
    
    # Build and train model
    n_features = X_train.shape[2]
    model = STransformer(
        window_size=window_size, 
        n_features=n_features, 
        d_model=64, 
        n_heads=4,
        ff_dim=256,
        n_blocks=2,
        dropout=0.2
    ).to(device)
    
    print(f"Model architecture:\n{model}")
    print(f"Number of parameters: {sum(p.numel() for p in model.parameters())}")
    
    # Train with early stopping
    print("Starting model training...")
    train_losses, val_losses = train_model(model, train_loader, val_loader, epochs=200, patience=20)
    
    # Plot training/validation loss curves
    plt.figure(figsize=(10, 6))
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss Curves')
    plt.legend()
    plt.tight_layout()
    plt.savefig('loss_curves.png', dpi=300, bbox_inches='tight')
    plt.close()
    
    # Evaluate on training set
    print("Evaluating on training set:")
    y_pred_train, y_true_train = evaluate_model(model, full_train_loader, target_scaler)
    
    # Evaluate on test set
    print("Evaluating on test set:")
    y_pred_test, y_true_test = evaluate_model(model, test_loader, target_scaler)
    
    # Create and evaluate baseline model
    print("Evaluating baseline model:")
    baseline_pred_train, baseline_pred_test = baseline_model(X_train, y_train, X_test)
    
    # If target was scaled, inverse transform baseline predictions
    if target_scaler:
        baseline_pred_train = target_scaler.inverse_transform(
            baseline_pred_train.reshape(-1, 1)).flatten()
        baseline_pred_test = target_scaler.inverse_transform(
            baseline_pred_test.reshape(-1, 1)).flatten()
    
    # Calculate baseline metrics
    baseline_rmse_train = np.sqrt(mean_squared_error(y_true_train, baseline_pred_train))
    baseline_r2_train = r2_score(y_true_train, baseline_pred_train)
    baseline_rmse_test = np.sqrt(mean_squared_error(y_true_test, baseline_pred_test))
    baseline_r2_test = r2_score(y_true_test, baseline_pred_test)
    
    print(f"Baseline Train RMSE: {baseline_rmse_train:.4f}, R: {baseline_r2_train:.4f}")
    print(f"Baseline Test RMSE: {baseline_rmse_test:.4f}, R: {baseline_r2_test:.4f}")
    
    # Plot results
    plot_results(train_dates, y_true_train, y_pred_train, test_dates, y_true_test, y_pred_test)
    
    # Save predictions to CSV
    train_results = pd.DataFrame({
        'date': train_dates,
        'observed': y_true_train,
        'predicted': y_pred_train,
        'baseline': baseline_pred_train
    })
    
    test_results = pd.DataFrame({
        'date': test_dates,
        'observed': y_true_test,
        'predicted': y_pred_test,
        'baseline': baseline_pred_test
    })
    
    train_results.to_csv('shoreline_predictions_train.csv', index=False)
    test_results.to_csv('shoreline_predictions_test.csv', index=False)
    
    print("Analysis complete. Results saved to CSV and plots generated.")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import random
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Conv1D, BatchNormalization, ReLU, Dropout, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import l2


# 0. Set random seeds for reproducibility
def set_seed(seed=42):
    np.random.seed(seed)
    random.seed(seed)
    tf.random.set_seed(seed)

set_seed(42)

# 1. Data Preparation
def load_and_preprocess_data(filepath):
    data = pd.read_csv(filepath, parse_dates=['dates'])

    # Create temporal features
    data['days_since_start'] = (data['dates'] - data['dates'].min()).dt.days
    data['month'] = data['dates'].dt.month
    data["Hs_ma_7"] = data['Hs'].rolling(7).mean()
    data["Hs_ma_30"] = data['Hs'].rolling(30).mean()
    data["Tp_ma_7"] = data['Tp'].rolling(7).mean()
    data["Tp_ma_30"] = data['Tp'].rolling(30).mean()
    data["Dir_ma_7"] = data['Dir'].rolling(7).mean()
    data["Dir_ma_30"] = data['Dir'].rolling(30).mean()
    data["Eng_ma_7"] = data['Eng'].rolling(7).mean()
    data["Eng_ma_30"] = data['Eng'].rolling(30).mean()

    data = data.dropna()

    train = data[data['dates'].dt.year <= 2020]
    test = data[data['dates'].dt.year > 2020]

    features = ['Hs', 'Tp', 'Dir', 'Eng', 'days_since_start', 'month', 'Hs_ma_7', 'Hs_ma_30', 
                'Tp_ma_7', 'Tp_ma_30', 'Dir_ma_7', 'Dir_ma_30', 'Eng_ma_7', 'Eng_ma_30']
    target = 'shore'

    scaler = StandardScaler()
    X_train = scaler.fit_transform(train[features])
    X_test = scaler.transform(test[features])
    y_train = train[target].values
    y_test = test[target].values
    train_dates = train['dates'].values
    test_dates = test['dates'].values

    return X_train, X_test, y_train, y_test, train_dates, test_dates, scaler

# 2. Create Sequences
def create_sequences(X, y, dates, window_size=30):
    X_seq, y_seq, date_seq = [], [], []
    for i in range(len(X) - window_size):
        X_seq.append(X[i:i+window_size])
        y_seq.append(y[i+window_size])
        date_seq.append(dates[i+window_size])
    return np.array(X_seq), np.array(y_seq), np.array(date_seq)

# 3. TCN Model
def build_tcn(input_shape, num_filters=64, kernel_size=3, dropout=0.4):
    inputs = Input(shape=input_shape)

    x = Conv1D(num_filters, kernel_size, padding='causal', dilation_rate=1, kernel_regularizer=l2(0.001))(inputs)
    x = BatchNormalization()(x)
    x = ReLU()(x)
    x = Dropout(dropout)(x)

    x = Conv1D(num_filters, kernel_size, padding='causal', dilation_rate=2, kernel_regularizer=l2(0.001))(x)
    x = BatchNormalization()(x)
    x = ReLU()(x)
    x = Dropout(dropout)(x)

    x = Conv1D(num_filters, kernel_size, padding='causal', dilation_rate=4, kernel_regularizer=l2(0.001))(x)
    x = BatchNormalization()(x)
    x = ReLU()(x)
    x = Dropout(dropout)(x)

    x = Flatten()(x)
    outputs = Dense(1)(x)

    return Model(inputs, outputs)

# 4. Main Execution
def main():
    WINDOW_SIZE = 30
    EPOCHS = 300
    BATCH_SIZE = 32

    X_train, X_test, y_train, y_test, train_dates, test_dates, scaler = load_and_preprocess_data(
        'C:/Users/amgh628/Desktop/DeepLearning/OTAMA_linear_smoothed.csv')

    X_train_seq, y_train_seq, train_date_seq = create_sequences(X_train, y_train, train_dates, WINDOW_SIZE)
    X_test_seq, y_test_seq, test_date_seq = create_sequences(X_test, y_test, test_dates, WINDOW_SIZE)

    #  Time-based validation split (80% train, 20% val)
    split_index = int(len(X_train_seq) * 0.8)
    X_train_final, X_val_final = X_train_seq[:split_index], X_train_seq[split_index:]
    y_train_final, y_val_final = y_train_seq[:split_index], y_train_seq[split_index:]

    model = build_tcn((WINDOW_SIZE, X_train_seq.shape[2]))
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

    early_stop = EarlyStopping(monitor='val_loss', patience=80, min_delta=0.0005, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=30, verbose=1)

    history = model.fit(X_train_final, y_train_final,
                        epochs=EPOCHS,
                        batch_size=BATCH_SIZE,
                        validation_data=(X_val_final, y_val_final),
                        callbacks=[early_stop, reduce_lr],
                        verbose=1)

    # Predict
    train_pred = model.predict(X_train_seq)
    test_pred = model.predict(X_test_seq)

    # Evaluation
    print("\nEvaluation Metrics (Test):")
    print(f"Test MAE: {mean_absolute_error(y_test_seq, test_pred):.2f}")
    print(f"Test RMSE: {np.sqrt(mean_squared_error(y_test_seq, test_pred)):.2f}")
    print(f"Test R: {r2_score(y_test_seq, test_pred):.2f}")

    # Plot
    plt.figure(figsize=(14, 6))
    plt.plot(train_date_seq, y_train_seq, label='Actual (Train)', color='blue', alpha=0.7)
    plt.plot(train_date_seq, train_pred, label='Predicted (Train)', color='cyan', linestyle='--')
    plt.plot(test_date_seq, y_test_seq, label='Actual (Test)', color='green', alpha=0.7)
    plt.plot(test_date_seq, test_pred, label='Predicted (Test)', color='red', linestyle='--')

    plt.title('Shoreline Position Prediction (1999-2025)')
    plt.xlabel('Year')
    plt.ylabel('Shoreline Position')
    plt.legend()
    plt.grid(True)
    plt.gca().xaxis.set_major_locator(mdates.YearLocator(4))
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    plt.gcf().autofmt_xdate()
    plt.savefig('shoreline_prediction.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    
    # Save results
    results = pd.DataFrame({
        'Date': np.concatenate([train_date_seq, test_date_seq]),
        'Actual': np.concatenate([y_train_seq, y_test_seq]),
        'Predicted': np.concatenate([train_pred.flatten(), test_pred.flatten()])
    })
    results.to_csv('prediction_OTAMA3.csv', index=False)
    print("Results saved to 'results.csv'.")

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import random
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


# 0. Set random seeds
def set_seed(seed=42):
    np.random.seed(seed)
    random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

set_seed(42)

# 1. Data Preparation
def load_and_preprocess_data(filepath):
    data = pd.read_csv(filepath, parse_dates=['dates'])

    data['days_since_start'] = (data['dates'] - data['dates'].min()).dt.days
    data['month'] = data['dates'].dt.month
    data["Hs_ma_7"] = data['Hs'].rolling(7).mean()
    data["Hs_ma_30"] = data['Hs'].rolling(30).mean()
    data["Tp_ma_7"] = data['Tp'].rolling(7).mean()
    data["Tp_ma_30"] = data['Tp'].rolling(30).mean()
    data["Dir_ma_7"] = data['Dir'].rolling(7).mean()
    data["Dir_ma_30"] = data['Dir'].rolling(30).mean()
    data["Eng_ma_7"] = data['Eng'].rolling(7).mean()
    data["Eng_ma_30"] = data['Eng'].rolling(30).mean()

    data = data.dropna()

    train = data[data['dates'].dt.year <= 2020]
    test = data[data['dates'].dt.year > 2020]

    features = ['Hs', 'Tp', 'Dir', 'Eng', 'days_since_start', 'month', 'Hs_ma_7', 'Hs_ma_30', 
                'Tp_ma_7', 'Tp_ma_30', 'Dir_ma_7', 'Dir_ma_30', 'Eng_ma_7', 'Eng_ma_30']
    target = 'shore'

    scaler = StandardScaler()
    X_train = scaler.fit_transform(train[features])
    X_test = scaler.transform(test[features])
    y_train = train[target].values
    y_test = test[target].values
    train_dates = train['dates'].values
    test_dates = test['dates'].values

    return X_train, X_test, y_train, y_test, train_dates, test_dates, scaler

# 2. Create Sequences
def create_sequences(X, y, dates, window_size=30):
    X_seq, y_seq, date_seq = [], [], []
    for i in range(len(X) - window_size):
        X_seq.append(X[i:i+window_size])
        y_seq.append(y[i+window_size])
        date_seq.append(dates[i+window_size])
    return np.array(X_seq), np.array(y_seq), np.array(date_seq)

# 3. TCN in PyTorch
class TCN(nn.Module):
    def __init__(self, input_dim, num_filters=64, kernel_size=3, dropout=0.4):
        super(TCN, self).__init__()

        def causal_conv(in_channels, out_channels, dilation):
            padding = (kernel_size - 1) * dilation
            return nn.Sequential(
                nn.ConstantPad1d((padding, 0), 0),  # causal padding
                nn.Conv1d(in_channels, out_channels, kernel_size, dilation=dilation),
                nn.BatchNorm1d(out_channels),
                nn.ReLU(),
                nn.Dropout(dropout)
            )

        self.tcn = nn.Sequential(
            causal_conv(input_dim, num_filters, dilation=1),
            causal_conv(num_filters, num_filters, dilation=2),
            causal_conv(num_filters, num_filters, dilation=4)
        )
        self.fc = nn.Linear(num_filters * 30, 1)  # 30 = window size

    def forward(self, x):
        x = x.permute(0, 2, 1)  # shape: (batch, channels, time)
        x = self.tcn(x)
        x = x.view(x.size(0), -1)  # flatten
        return self.fc(x)

# 4. Main Training Routine
def main():
    WINDOW_SIZE = 30
    EPOCHS = 300
    BATCH_SIZE = 32
    LEARNING_RATE = 0.001

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    X_train, X_test, y_train, y_test, train_dates, test_dates, scaler = load_and_preprocess_data(
        'C:/Users/amgh628/Desktop/DeepLearning/OTAMA_linear_smoothed.csv')

    X_train_seq, y_train_seq, train_date_seq = create_sequences(X_train, y_train, train_dates, WINDOW_SIZE)
    X_test_seq, y_test_seq, test_date_seq = create_sequences(X_test, y_test, test_dates, WINDOW_SIZE)

    split_index = int(len(X_train_seq) * 0.8)
    X_train_final = torch.tensor(X_train_seq[:split_index], dtype=torch.float32)
    y_train_final = torch.tensor(y_train_seq[:split_index], dtype=torch.float32).unsqueeze(1)
    X_val_final = torch.tensor(X_train_seq[split_index:], dtype=torch.float32)
    y_val_final = torch.tensor(y_train_seq[split_index:], dtype=torch.float32).unsqueeze(1)

    X_test_tensor = torch.tensor(X_test_seq, dtype=torch.float32)
    y_test_tensor = torch.tensor(y_test_seq, dtype=torch.float32).unsqueeze(1)

    train_dataset = TensorDataset(X_train_final, y_train_final)
    val_dataset = TensorDataset(X_val_final, y_val_final)
    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE)

    model = TCN(input_dim=X_train_seq.shape[2]).to(device)
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)

    best_val_loss = float('inf')
    patience, patience_counter = 80, 0

    for epoch in range(EPOCHS):
        model.train()
        for xb, yb in train_loader:
            xb, yb = xb.to(device), yb.to(device)
            optimizer.zero_grad()
            pred = model(xb)
            loss = criterion(pred, yb)
            loss.backward()
            optimizer.step()

        model.eval()
        with torch.no_grad():
            val_preds = model(X_val_final.to(device))
            val_loss = criterion(val_preds, y_val_final.to(device)).item()

        print(f"Epoch {epoch+1}/{EPOCHS} - Val Loss: {val_loss:.4f}")

        if val_loss < best_val_loss - 0.0005:
            best_val_loss = val_loss
            best_model_state = model.state_dict()
            patience_counter = 0
        else:
            patience_counter += 1
            if patience_counter >= patience:
                print("Early stopping.")
                break

    # Load best model
    model.load_state_dict(best_model_state)

    # Evaluate
    model.eval()
    with torch.no_grad():
        train_preds = model(torch.tensor(X_train_seq, dtype=torch.float32).to(device)).cpu().numpy()
        test_preds = model(X_test_tensor.to(device)).cpu().numpy()

    print("\nEvaluation Metrics (Test):")
    print(f"Test MAE: {mean_absolute_error(y_test_seq, test_preds):.2f}")
    print(f"Test RMSE: {np.sqrt(mean_squared_error(y_test_seq, test_preds)):.2f}")
    print(f"Test R: {r2_score(y_test_seq, test_preds):.2f}")

    # Plot
    plt.figure(figsize=(14, 6))
    plt.plot(train_date_seq, y_train_seq, label='Actual (Train)', color='blue', alpha=0.7)
    plt.plot(train_date_seq, train_preds, label='Predicted (Train)', color='cyan', linestyle='--')
    plt.plot(test_date_seq, y_test_seq, label='Actual (Test)', color='green', alpha=0.7)
    plt.plot(test_date_seq, test_preds, label='Predicted (Test)', color='red', linestyle='--')

    plt.title('Shoreline Position Prediction (19992025)')
    plt.xlabel('Year')
    plt.ylabel('Shoreline Position')
    plt.legend()
    plt.grid(True)
    plt.gca().xaxis.set_major_locator(mdates.YearLocator(4))
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    plt.gcf().autofmt_xdate()
    plt.savefig('shoreline_prediction_pytorch.png', dpi=300, bbox_inches='tight')
    plt.show()

    # Save results
    results = pd.DataFrame({
        'Date': np.concatenate([train_date_seq, test_date_seq]),
        'Actual': np.concatenate([y_train_seq, y_test_seq]),
        'Predicted': np.concatenate([train_preds.flatten(), test_preds.flatten()])
    })
    results.to_csv('prediction_OTAMA3_pytorch.csv', index=False)
    print("Results saved to 'prediction_OTAMA3_pytorch.csv'.")

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import random
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Conv1D, BatchNormalization, ReLU, Dropout, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import l2


# 0. Set random seeds for reproducibility
def set_seed(seed=42):
    np.random.seed(seed)
    random.seed(seed)
    tf.random.set_seed(seed)

set_seed(42)

# 1. Data Preparation
def load_and_preprocess_data(filepath):
    data = pd.read_csv(filepath, parse_dates=['dates'])

    # Create temporal features
    data['days_since_start'] = (data['dates'] - data['dates'].min()).dt.days
    data['month'] = data['dates'].dt.month
    data["Hs_ma_7"] = data['Hs'].rolling(7).mean()
    data["Hs_ma_30"] = data['Hs'].rolling(30).mean()
    data["Tp_ma_7"] = data['Tp'].rolling(7).mean()
    data["Tp_ma_30"] = data['Tp'].rolling(30).mean()
    data["Dir_ma_7"] = data['Dir'].rolling(7).mean()
    data["Dir_ma_30"] = data['Dir'].rolling(30).mean()
    data["Eng_ma_7"] = data['Eng'].rolling(7).mean()
    data["Eng_ma_30"] = data['Eng'].rolling(30).mean()

    data = data.dropna()

    train = data[data['dates'].dt.year <= 2020]
    test = data[data['dates'].dt.year > 2020]

    features = ['Hs', 'Tp', 'Dir', 'Eng', 'days_since_start', 'month', 'Hs_ma_7', 'Hs_ma_30', 
                'Tp_ma_7', 'Tp_ma_30', 'Dir_ma_7', 'Dir_ma_30', 'Eng_ma_7', 'Eng_ma_30']
    target = 'shore'

    scaler = StandardScaler()
    X_train = scaler.fit_transform(train[features])
    X_test = scaler.transform(test[features])
    y_train = train[target].values
    y_test = test[target].values
    train_dates = train['dates'].values
    test_dates = test['dates'].values

    return X_train, X_test, y_train, y_test, train_dates, test_dates, scaler

# 2. Create Sequences
def build_stronger_tcn(input_shape, num_filters=128, kernel_size=3, dropout=0.2):
    inputs = Input(shape=input_shape)
    
    # First block - dilation rate 1
    x = Conv1D(num_filters, kernel_size, padding='causal', dilation_rate=1, 
               kernel_regularizer=l2(0.0001))(inputs)
    x = BatchNormalization()(x)
    x = ReLU()(x)
    x = Dropout(dropout)(x)
    
    # Second block - dilation rate 2
    res_x = x  # Store for residual connection
    x = Conv1D(num_filters, kernel_size, padding='causal', dilation_rate=2, 
               kernel_regularizer=l2(0.0001))(x)
    x = BatchNormalization()(x)
    x = ReLU()(x)
    x = Dropout(dropout)(x)
    # Add residual connection
    x = Conv1D(num_filters, 1, padding='same')(res_x) + x  # 1x1 conv for dimension matching
    
    # Third block - dilation rate 4
    res_x = x  # Store for residual connection
    x = Conv1D(num_filters, kernel_size, padding='causal', dilation_rate=4, 
               kernel_regularizer=l2(0.0001))(x)
    x = BatchNormalization()(x)
    x = ReLU()(x)
    x = Dropout(dropout)(x)
    # Add residual connection
    x = x + res_x
    
    # Fourth block - dilation rate 8 (new)
    res_x = x
    x = Conv1D(num_filters*2, kernel_size, padding='causal', dilation_rate=8, 
               kernel_regularizer=l2(0.0001))(x)
    x = BatchNormalization()(x)
    x = ReLU()(x)
    x = Dropout(dropout)(x)
    # Add residual connection (with dimension matching)
    x = x + Conv1D(num_filters*2, 1, padding='same')(res_x)
    
    # Fifth block - dilation rate 16 (new)
    res_x = x
    x = Conv1D(num_filters*2, kernel_size, padding='causal', dilation_rate=16, 
               kernel_regularizer=l2(0.0001))(x)
    x = BatchNormalization()(x)
    x = ReLU()(x)
    x = Dropout(dropout)(x)
    # Add residual connection
    x = x + res_x
    
    # Output layers
    x = Conv1D(num_filters, 1, padding='same')(x)  # Bottleneck layer
    x = BatchNormalization()(x)
    x = ReLU()(x)
    x = Flatten()(x)
    x = Dense(64, activation='relu', kernel_regularizer=l2(0.0001))(x)  # Add dense layer
    outputs = Dense(1)(x)
    
    return Model(inputs, outputs)

# 4. Main Execution
def main():
    WINDOW_SIZE = 30
    EPOCHS = 300
    BATCH_SIZE = 32

    X_train, X_test, y_train, y_test, train_dates, test_dates, scaler = load_and_preprocess_data(
        'C:/Users/amgh628/Desktop/DeepLearning/OTAMA_linear_smoothed.csv')

    X_train_seq, y_train_seq, train_date_seq = create_sequences(X_train, y_train, train_dates, WINDOW_SIZE)
    X_test_seq, y_test_seq, test_date_seq = create_sequences(X_test, y_test, test_dates, WINDOW_SIZE)

    #  Time-based validation split (80% train, 20% val)
    split_index = int(len(X_train_seq) * 0.8)
    X_train_final, X_val_final = X_train_seq[:split_index], X_train_seq[split_index:]
    y_train_final, y_val_final = y_train_seq[:split_index], y_train_seq[split_index:]

    model = build_tcn((WINDOW_SIZE, X_train_seq.shape[2]))
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

    early_stop = EarlyStopping(monitor='val_loss', patience=80, min_delta=0.0005, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=30, verbose=1)

    history = model.fit(X_train_final, y_train_final,
                        epochs=EPOCHS,
                        batch_size=BATCH_SIZE,
                        validation_data=(X_val_final, y_val_final),
                        callbacks=[early_stop, reduce_lr],
                        verbose=1)

    # Predict
    train_pred = model.predict(X_train_seq)
    test_pred = model.predict(X_test_seq)

    # Evaluation
    print("\nEvaluation Metrics (Test):")
    print(f"Test MAE: {mean_absolute_error(y_test_seq, test_pred):.2f}")
    print(f"Test RMSE: {np.sqrt(mean_squared_error(y_test_seq, test_pred)):.2f}")
    print(f"Test R: {r2_score(y_test_seq, test_pred):.2f}")

    # Plot
    plt.figure(figsize=(14, 6))
    plt.plot(train_date_seq, y_train_seq, label='Actual (Train)', color='blue', alpha=0.7)
    plt.plot(train_date_seq, train_pred, label='Predicted (Train)', color='cyan', linestyle='--')
    plt.plot(test_date_seq, y_test_seq, label='Actual (Test)', color='green', alpha=0.7)
    plt.plot(test_date_seq, test_pred, label='Predicted (Test)', color='red', linestyle='--')

    plt.title('Shoreline Position Prediction (1999-2025)')
    plt.xlabel('Year')
    plt.ylabel('Shoreline Position')
    plt.legend()
    plt.grid(True)
    plt.gca().xaxis.set_major_locator(mdates.YearLocator(4))
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    plt.gcf().autofmt_xdate()
    plt.savefig('shoreline_prediction.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    
    # Save results
    results = pd.DataFrame({
        'Date': np.concatenate([train_date_seq, test_date_seq]),
        'Actual': np.concatenate([y_train_seq, y_test_seq]),
        'Predicted': np.concatenate([train_pred.flatten(), test_pred.flatten()])
    })
    results.to_csv('prediction_OTAMA4.csv', index=False)
    print("Results saved to 'results.csv'.")

if __name__ == "__main__":
    main()
```
